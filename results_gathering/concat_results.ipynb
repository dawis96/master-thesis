```python
import pandas as pd
import os
import glob
```

```python
# Find all csv
extension = 'csv'
csv_list = glob.glob('*.{}'.format(extension))
```

```python
df_list = []
for csv in csv_list:
    df_temp = pd.read_csv(csv)
    df_list.append(df_temp)
```

```python
df = pd.concat(df_list, ignore_index=True)
```
